```python
import pandas as pd
```

```python
#Leitura da Planilha STG_FNT_ITT
df_FNT_ITT = pd.read_excel ('STG_FNT_ITT.xlsx', delimiter = ';')
```

```python
#Remoção de dados duplicados
def del_duplicates (x):
    x = x.drop_duplicates()
    return x
del_duplicates (df_FNT_ITT)
```

```python
#Completar zeros à esquerda do NUM_CNPJ
df_FNT_ITT['NUM_CNPJ'] = df_FNT_ITT['NUM_CNPJ'].astype(str)
width = 11
df_FNT_ITT["NUM_CNPJ"]= df_FNT_ITT["NUM_CNPJ"].str.zfill(width) 
print (df_FNT_ITT["NUM_CNPJ"])

```
